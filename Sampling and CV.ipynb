```python
pip install imblearn
```

```python
x_smt, y_smt = smt.fit_resample(x,y)
```

```python
y_smt.value_counts()
```

```python
x_smt_train.shape
```

```python
x_smt_test.shape
```

```python
score_dtc_smt = dtc_smt.score(x_smt_test,y_smt_test)
```

```python
score_dtc_smt
```

### Random Forest Classifier

```python
rfc_smt = RandomForestClassifier()
rfc_smt.fit(x_smt_train,y_smt_train)
```

```python
score_rfc_smt = rfc_smt.score(x_smt_test,y_smt_test)
score_rfc_smt
```

### XGBoost Classifier

```python
xgbc_smt = XGBClassifier()
xgbc_smt.fit(x_smt_train,y_smt_train)
```

```python
score_xgbc_smt = xgbc_smt.score(x_smt_test,y_smt_test)
score_xgbc_smt
```

### Ada Boost classifier

```python
abc_smt = AdaBoostClassifier()
abc_smt.fit(x_smt_train,y_smt_train)
```

```python
score_abc_smt = abc_smt.score(x_smt_test,y_smt_test)
score_abc_smt
```

### KNN Classifier

```python
knnc_smt = KNeighborsClassifier()
knnc_smt.fit(x_smt_train,y_smt_train)
```

```python
score_knnc_smt = knnc_smt.score(x_smt_test,y_smt_test)
score_knnc_smt
```

### Logistic Regression

```python
lrc_smt = LogisticRegression()
lrc_smt.fit(x_smt_train,y_smt_train)
```

```python
df_smt = pd.DataFrame([['Decision Tree (SMOTETomek)',score_dtc_smt],['Random Forest (SMOTETomek)',score_rfc_smt],['XGBoost (SMOTETomek)',score_xgbc_smt],['Ada Boost (SMOTETomek)',score_abc_smt],['K-Nearest Neighbor (SMOTETomek)',score_knnc_smt],['Logistic Regression (SMOTETomek)',score_lrc_smt]],columns=['Classifier', 'Accuracy'])
df_smt
```
